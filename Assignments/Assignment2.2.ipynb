# Problem 1 [2p] Naive Bayes Classifier

The Bayes' theorem allows us to construct a classifier in which we
model how the data is generated. Here we will describe a
simple and popular example of such a classifier called the naive
Bayes classifier.  Despite its simplicity It is quite effective for
classification of text documents (e.g. as spam and non-spam).

Let a document be a sequence of tokens, which usually are words, but can also be characters: $D=W_1,W_2,\ldots,W_n$ 
We will model generation of text documents as a two-stage process.
First, document category $C_j$ is drawn at random with probability
$p(C_j)$, also called the *a priori* probability.
To define the class-conditional probability
$p(D|C_j)$, we will make a simplifying (naive)
assumption, that every token in the document is drawn independently at
random with probability $p(W_i|C)$:

\begin{equation*}
p(D|C_j) = p(W_1,W_2,\ldots,W_n | C_j) \approx p(W_1|C_j)p(W_2|C_j)\ldots p(W_n|C_j).
\end{equation*}

To infer the class of a document we apply the Bayes theorem:
\begin{equation*}
p(C_j|D) = \frac{p(D|C_j)p(C_j)}{p(D)} = \frac{p(C_j)p(W_1|C_j)p(W_2|C_j)\ldots p(W_n|C_j)}{p(D)}.
\end{equation*}
Please note that since we assumed only a finite number of classes,
we can compute the term $p(D)$ by making sure that the *a
posteriori probabilities* $p(C_j|D)$ sum to $1$ over all classes.

```python
import re

def naive_bayes(sent, langs, df):
    """Returns the most probable language of a sentence"""

    # Try working with log-probabilities.
    # to prevent taking log(0) you can e.g. add a very small amount (1e-100)
    # to each tabulated frequency.
    df_log = np.log(df + 1e-100)

    # normalize the sentence: remove spaces and punctuations, take lower case
    sent = re.sub(r'[\s!"#$%&\'()*+,-\./:;<=>?@[\\\]^_`{|}~]', '', sent.lower())
    print(sent)

    log_probs = {}
    for lang in langs:
        log_probs[lang] = df_log.loc[list(sent), lang].sum()
        # log_probs[lang] = df_log.loc[list(sent), lang].mean() # use mean so values do not depend on sentence length # hmm, not this time, it yields worse results for some reason
    
    # TODO compute language probabilitie and order from most to least probable

    max_ll = max(log_probs.values())
    probs = { k: np.exp(v - max_ll) for k, v in log_probs.items() }
    likelihood_sum = sum(sorted(probs.values()))
    probs = dict(sorted( # dict is ordered by default in Python >= 3.7
        map(
            lambda k_v: (k_v[0], k_v[1] / likelihood_sum), # normalize likelihoods so they sum to 1
            probs.items()
        ),
        key=lambda k_v: -k_v[1]
        )
    )
    return probs
```

```python
#
# Implement the Rosenbrock function
#

def rosenbrock_v(x):
    """Returns the value of Rosenbrock's function at x"""
    x, y = x
    a, b = (1.0, 100.0)
    return np.square(a - x) + b * np.square(y - np.square(x))


def rosenbrock(x):
    """Returns the value of rosenbrock's function and its gradient at x
    """
    val = rosenbrock_v(x)
    x, y = x
    a, b = (1.0, 100.0)
    # Gradient should be np.array
    dVdX = np.array((
        2.0 * (x - a) - 4.0 * b * x * (y - np.square(x)),
        2.0 * b * (y - np.square(x))
    ))
    return [val, dVdX]


#
# Feel free to add your own test points.
#
for test_point in [[0.0, 0.0], [1.0, 1.0], [0.5, 1.0], [1.0, 0.5], [0.0, 1.0]]:
    assert check_gradient(rosenbrock, np.array(test_point), prec=1e-5)
```

# Problem 3 (Quantile Regression) [4p]

  The least squares method results in estimates that approximate
  the conditional mean of the response variable given certain values
  of the predictor variables.  However, for many applications we are interested
  in a median or other percentile rather than the mean of the response
  variable. An approximation of a percentile can be achieved
  indirectly by using the least squares method to fit a model,
  assuming a gaussian distribution on residuals of this model and
  calculating the required percentile of the gaussian distribution.
  The problems with this commonly used technique appear when the
  distribution of residuals does not follow a gaussian distribution.

  The quantile regression
  (https://en.wikipedia.org/wiki/Quantile_regression) aims at
  directly estimating a value of the conditional percentile of the
  response variable. It is often use in e.g. sales forecasting, where
  we are interested in e.g. keeping enough items to have a 90% chance
  to cover the demand. In quantile regression, instead of minimizing the mean squared error,
  quantile regression minimizes a different cost function namely:
  \begin{equation}
  f_{\tau} = \left\{
    \begin{array}{rl}
      \tau\cdot x & \text{if } x \geq 0,\\
      -(1-\tau)\cdot x & \text{if } x < 0,
    \end{array} \right.
  \end{equation}
  where $\tau$ is the precentile of interest.

    work well in practice.

### 3.1. Answer

```python
def powers_of_X(X, degree):
    powers = np.arange(degree + 1).reshape(1, -1)
    return X ** powers

def eval_poly(X, Theta):
    XP = powers_of_X(X, len(Theta) - 1)  # len(Theta) x N
    Y = XP @ Theta
    return Y 
```

```python
def poly_fit(data, degree=1, alpha=0):
    "Fit a polynomial of a given degree and an optional weight decay parameter alpha"
    X = powers_of_X(data[0], degree)  # Matrix N x d
    Y = data[1].reshape(-1, 1)  # Matrix N x 1
    d = X.shape[1]
    Theta = np.linalg.solve(
        X.T @ X + np.diag(np.repeat(alpha, d)),
        X.T @ Y
        )
    return Theta
```

```python
lsq_Theta = poly_fit((X, Y))
lsq_Theta
```

```python
predicted_Y = eval_poly(X, lsq_Theta)
residuals = Y - predicted_Y

# Fit residuals to normal dist. and get quants
mean, std = sstats.norm.fit(residuals)
q1, q2 = sstats.norm.ppf([.25, .75], mean, std)
original_q1, original_q2 = q1, q2 # save for quantile regression plot

# make a plot
fig, ax = plt.subplots(1, 1)
ax.set_title('Normalized histogram of residuals and the red line that do not look like PDF for this histogram')
ax.set_xlabel('Residual')
ax.set_ylabel('Residual count')
ax.set_xlim(0, 400)

# plot the histogram
plt.hist(residuals, bins=2000, density=True) # use density=False for histogram to have residual count on Y axis

# plot PDF with parameters from fitting part above
pdf_xs = np.linspace(0, 400, num=100)
pdf_ys = sstats.norm.pdf(pdf_xs, loc=mean, scale=std)
plt.plot(pdf_xs, pdf_ys, color='r') # let's scale PDF so we can see anything
```

```python
print(
    f"Theta from Least squares: {lsq_Theta}",
    f"Fraction of errs below 25 perc: {(residuals < q1).mean()}, above 75 perc: {(q2 < residuals).mean()}",
    f"Quantiles of normal distribution: {[q1, q2]}",
    sep='\n'
)
```

```python
def quant_reg(tau, X=X, Y=Y): # this is meant for linear case only ofc
    def quant_reg_loss_val_grad(Theta):
        """Returns the value of rosenbrock's function and its gradient at x
        """
        predicted_Y = X @ Theta
        residuals = Y - predicted_Y[:, np.newaxis]
        cond = 0.0 <= residuals
        ic.enable()

        val = np.where(
            cond,
            tau * residuals,
            (tau - 1.0) * residuals
        ).mean()

        grad = np.where( 
            cond,
            -tau * X,
            (1.0 - tau) * X
        ).mean(axis=0)
        return [val, grad]

    x_start = [1.0, 1.0]
    Theta, _, _ = sopt.fmin_l_bfgs_b(
        quant_reg_loss_val_grad, x_start
    )

    predicted_Y = X @ Theta
    residuals = Y - predicted_Y[:, np.newaxis]
    neg_errs_fract = (residuals < 0).mean()

    print(
        f'Quant {tau} theta {Theta}',
        f'Fraction of negative errs {neg_errs_fract}: (should be {tau})',
        sep='\n'
        )

    return Theta
```

```python
Theta_25 = quant_reg(0.25)
```

```python
Theta_50 = quant_reg(0.5)
```

```python
Theta_75 = quant_reg(0.75)
```

```python
# make a plot
fig, ax = plt.subplots(1, 1)
ax.set_xlim(None, 250)
ax.set_ylim(None, 3000)
data = data[data.price.between(0, 4000)]
plt.scatter(data.area, data.price, alpha=0.25)

def plot_theta_line(Theta, color, label):
    c, a = Theta
    xs = np.linspace(0, 400, num=5)
    ys = xs * a + c
    plt.plot(xs, ys, color=color, label=label)

lsq_Theta_25 = lsq_Theta + np.array([original_q1, 0.0])[:, np.newaxis] # add error quantile value to to c term
lsq_Theta_75 = lsq_Theta + np.array([original_q2, 0.0])[:, np.newaxis] # add error quantile value to to c term

plot_theta_line(lsq_Theta, 'b', label='LSQ') # gets overlayed by Theta_75
plot_theta_line(lsq_Theta_25, 'orange', label='LSQ25')
plot_theta_line(lsq_Theta_75, 'g', label='LSQ75')
plot_theta_line(Theta_25, 'r', label='Q25')
plot_theta_line(Theta_50, 'purple', label='Q50')
plot_theta_line(Theta_75, 'brown', label='Q75')

plt.legend(loc='upper left')
```
